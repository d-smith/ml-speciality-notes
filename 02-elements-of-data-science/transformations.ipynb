```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

```python
df = pd.DataFrame({'a':np.random.rand(5),'b':np.random.rand(5)})
cube = PolynomialFeatures(degree=3)
cube_features = cube.fit_transform(df)
cube_features
cols = ['ones','a','b','a^2','ab','b^2','a^3','ba^2','ab^2','b^3']
df_cube=pd.DataFrame(cube_features,columns=cols)
df_cube
```
